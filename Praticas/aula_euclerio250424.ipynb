### Aplicado em DataFrame

```python
df2 = df_aux[['Unit_Cost', 'Unit_Price']]
```

```python
varejo = df2['Unit_Price']
atacado = df_aux['Unit_Cost']
```

```python
varejo - atacado #lucro
```

```python
def generate_profit(df):
    response = df['Unit_Price'] - df['Unit_Cost']
    return response
```

```python
df_aux['Lucro'] = df_aux.apply(generate_profit, axis = 1) #o apply salva as alterações
df_aux['Lucro']
```

```python
df_aux #ganho da coluna lucro
```
