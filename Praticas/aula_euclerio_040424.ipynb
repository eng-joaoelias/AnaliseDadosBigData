```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
capital = 2000
bet_size = 1

prob_win = 0.5
prob_loss = (1 - prob_win)

win_reward = 100
loss_reward = -100
```

```python
fig, ax = plt.subplots(figsize = (12,6))
last_response = []

for j in range(100):
    equity_curve = [capital]

    for num in range(10000):
        if prob_loss < random.random():
            equity_curve += [equity_curve[-1] + win_reward * bet_size]
        else:
            equity_curve += [equity_curve[-1] + loss_reward * bet_size]

    last_response += [equity_curve[-1] - capital]
    ax.plot(equity_curve)
```

```python
soma = sum(last_response) / len(last_response)
print(soma)
```

```python
capital = 1000
bet_size = 10
bet_size_start = 10

prob_win = 0.65
prob_loss = (1 - prob_win)

win_reward = 0.8
loss_reward = -1.4
```

```python
fig, ax = plt.subplots(figsize = (12,6))
quebra_banca = 0
last_response = []

for j in range(100):
    equity_curve = [capital]
    for num in range(10000):
        if prob_loss < random.random():
            equity_curve += [equity_curve[-1] + win_reward * bet_size]
            bet_size = bet_size_start
        else:
            equity_curve += [equity_curve[-1] + loss_reward * bet_size]
            bet_size *= 2
        if equity_curve[-1] <= 0:
            quebra_banca += 1

    last_response += [equity_curve[-1] - capital]
    ax.plot(equity_curve)

plt.show()
```
