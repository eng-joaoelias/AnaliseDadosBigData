# 5. Crie uma nova coluna para representar somente o mês e o ano (mm-AAAA), usando a coluna data final como referencia

```python
df_backup['MES_ANO_FINAL'] = df_backup['DATA FINAL'].dt.to_period('M')
```

```python
df_backup['MES_ANO_FINAL']
```

```python
df_backup['ANO-MES'] = (df_backup['DATA FINAL'].apply(lambda x: "{}".format(x.year)) + df_backup['DATA FINAL'].apply(lambda x: "-{:02d}".format(x.month)))
```

# 8. Preço médio de revenda da gasolina agosto/2008

```python
ago_2008 = dados_gasolina[dados_gasolina['MES_ANO_FINAL'] == pd.Period('2008-08')]
```

```python
media_revenda = ago_2008['PREÇO MÉDIO REVENDA'].mean().round(3)
print('O valor médio de revenda da gasolina em 08-2008 eh {}'.format(media_revenda))
```

# 9. Preço médio de revenda da gasolina em maio/2014 em São Paulo

```python
maio_2014 = dados_gasolina[dados_gasolina['MES_ANO_FINAL'] == pd.Period('2014-05')]
```

```python
sao_paulo = maio_2014[maio_2014['ESTADO'] == 'SAO PAULO']
```

```python
media_sp_052014 = sao_paulo['PREÇO MÉDIO REVENDA'].mean().round(3)
print('O preço médio de revenda da gasolina em 05/2014 em SP é {}'.format(media_sp_052014))
```

# 10. Quais estados a gasolina ultrapassou 5 reais? Quando isso ocorreu?

```python
maior_que_5_reais = dados_gasolina[dados_gasolina['PREÇO MÉDIO REVENDA'] >= 5][['ESTADO', 'ANO-MES', 'PREÇO MÉDIO REVENDA']]
```

```python
print('Estados onde a gasolina ultrapassou os 5 reais:\n{}'.format(maior_que_5_reais.value_counts()))
```
