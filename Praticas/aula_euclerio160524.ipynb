```python
import requests as req
from bs4 import BeautifulSoup
```

```python
response = req.get("https://free-proxy-list.net/") #operacao de get para o site que tem um conjunto de proxy
```

```python
html = response.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
table = soup.findAll(class_='table-responsive fpl-list')
```

```python
proxies = []
if table:
    #Iterar sobre cada linha da tabela
    for row in table: #Pular o cabecalho usando [1:]
        cells = row.find_all('td')
        if cells: # Verificar se c√©lulas foram encontradas
            for proxy in cells:
                print(proxy)
                proxy = {
                    'IP Address': cells[0].text.strip(),
                    'Port': cells[1].text.strip(),
                    'Code': cells[2].text.strip(),
                    'Country': cells[3].text.strip(),
                    'Anonymity': cells[4].text.strip(),
                    'Google': cells[5].text.strip(),
                    'HTTPS': cells[6].text.strip(),
                    'Last Checked': cells[7].text.strip(),
                }
                proxies.append(proxy)
```

```python
for proxy in proxies:
    print(proxy)
```
